```python
import numpy as np
from tensorflow.keras.layers import Dense,Flatten,Dropout,Input,Conv2D,MaxPool2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import img_to_array,load_img
from tensorflow.keras.utils import to_categorical
import os
```

```python
dir="../input/gtsrb-german-traffic-sign/Train/"
categories=os.listdir(dir)
```

```python
train_data,train_labels=[],[]

for category in categories:
    path=os.path.join(dir,category)
    for img in os.listdir(path):
        img_path=os.path.join(path,img)
        image=load_img(img_path,target_size=(30,30))
        image=img_to_array(image)

        train_data.append(image)
        train_labels.append(category)
```

```python
test_data,test_labels=[],[]

for category in categories:
    path=os.path.join(dir,category)
    for img in os.listdir(path):
        img_path=os.path.join(path,img)
        image=load_img(img_path,target_size=(30,30))
        image=img_to_array(image)

        test_data.append(image)
        test_labels.append(category)
```

```python
train_labels=to_categorical(train_labels,43)

train_data=np.array(train_data,dtype="float32")
train_labels=np.array(train_labels)
```

```python
test_labels=to_categorical(test_labels,43)

test_data=np.array(test_data,dtype="float32")
test_labels=np.array(test_labels)
```

```python
test_labels[1]
```

```python
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=(30,30,3)))
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(300, activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax'))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
model.fit(train_data,train_labels,batch_size=32,validation_data=(test_data,test_labels),epochs=30)
```

```python
model.save("traffic_model.h5")
```
