```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set=pd.read_csv("kaggle_diabetes.csv")
data_set.isna().sum()
```

```python
sns.pairplot(data_set,hue="Outcome")
plt.show()
```

```python
x=data_set.iloc[:,:-1].values
y=data_set.iloc[:,-1].values
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_classifier=LogisticRegression()
log_classifier.fit(x_train,y_train)
```

```python
plt.plot(sc_x.inverse_transform(x_train),y_train,'r')
plt.plot(sc_x.inverse_transform(x_train),log_classifier.predict(x_train),"b")
plt.title("Train Prediction graph")
plt.show()
```

```python
plt.plot(sc_x.inverse_transform(x_test),y_test,'r')
plt.plot(sc_x.inverse_transform(x_test),log_classifier.predict(x_test),"b")
plt.title("Test Prediction graph")
plt.show()
```

```python
print("Train Dataset Score: ",log_classifier.score(x_train,y_train))
print("Test Dataset Score: ",log_classifier.score(x_test,y_test))
```

```python
y_pred_train=log_classifier.predict(x_train)
y_pred_test=log_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
print("Training Confusion matrix: \n",confusion_matrix(y_train,y_pred_train))
print("Test Confusion matrix: \n",confusion_matrix(y_test,y_pred_test))
```

## svc(Kernel)

```python
from sklearn.svm import SVC
svc_classifier=SVC(kernel="rbf",random_state=42)
svc_classifier.fit(x_train,y_train)
```

```python
plt.plot(sc_x.inverse_transform(x_train),y_train,'r')
plt.plot(sc_x.inverse_transform(x_train),svc_classifier.predict(x_train),"b")
plt.title("Train Prediction graph")
plt.show()
```

```python
plt.plot(sc_x.inverse_transform(x_test),y_test,'r')
plt.plot(sc_x.inverse_transform(x_test),svc_classifier.predict(x_test),"b")
plt.title("Test Prediction graph")
plt.show()
```

```python
print("Train Dataset Score: ",svc_classifier.score(x_train,y_train))
print("Test Dataset Score: ",svc_classifier.score(x_test,y_test))
```

```python
y_pred_train=svc_classifier.predict(x_train)
y_pred_test=svc_classifier.predict(x_test)

from sklearn.metrics import confusion_matrix
print("Training Confusion matrix: \n",confusion_matrix(y_train,y_pred_train))
print("Test Confusion matrix: \n",confusion_matrix(y_test,y_pred_test))
```
